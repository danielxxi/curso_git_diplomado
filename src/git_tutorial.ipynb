## 1. ¿Qué es Git?

**Git** es un **sistema de control de versiones distribuido** (DVCS) creado por  
**Linus Torvalds** en 2005 para gestionar el kernel de Linux.

### ¿Por qué necesitamos control de versiones?

Imagina trabajar en un proyecto sin Git:

```
proyecto_v1.zip
proyecto_v2_final.zip
proyecto_v2_final_FINAL.zip
proyecto_v2_USAESTE.zip          ← ¿Cuál es el bueno?
```

Con Git, cada cambio queda registrado con:
- **Qué** cambió (diff)
- **Quién** lo cambió (autor)
- **Cuándo** cambió (timestamp)
- **Por qué** cambió (mensaje de commit)

### Git vs Otros VCS

| Característica | Git | SVN | CVS |
|----------------|-----|-----|-----|
| Tipo | Distribuido | Centralizado | Centralizado |
| Velocidad | ⚡ Muy rápido | Lento | Lento |
| Trabajo offline | ✅ Sí | ❌ No | ❌ No |
| Branching | ✅ Ligero | ⚠️ Costoso | ❌ Difícil |
| Popularidad | #1 Mundial | Legado | Obsoleto |

### Git ≠ GitHub

| Git | GitHub |
|-----|--------|
| Software de control de versiones | Plataforma web para alojar repositorios Git |
| Corre en tu máquina (local) | Corre en la nube (remoto) |
| Open Source, gratis | Producto comercial (con tier gratuito) |
| Creado en 2005 | Fundado en 2008 |
| Alternativas: ninguna | Alternativas: GitLab, Bitbucket, Gitea |

### Configuración Global (¡Hacer esto primero!)

La configuración de Git se guarda en `~/.gitconfig` (global) o `.git/config` (por proyecto).

```bash

git config --global user.name  "Tu Nombre"

git config --global core.editor "code --wait"   # VS Code
git config --global core.editor "vim"            # Vim
git config --global core.editor "nano"           # Nano

# ─── Branch principal por defecto (Git ≥ 2.28) ───────────
git config --global init.defaultBranch main

# ─── Colores en la terminal ───────────────────────────────
git config --global color.ui auto

# ─── Alias útiles ─────────────────────────────────────────
git config --global alias.st   status
git config --global alias.co   checkout
git config --global alias.br   branch
git config --global alias.lg   "log --oneline --graph --decorate --all"
git config --global alias.last "log -1 HEAD"
git config --global alias.unstage "reset HEAD --"
```

### `git clone` — Clonar un repositorio existente

Descarga un repositorio completo (con todo su historial) a tu máquina local.


## 4. Flujo Básico de Trabajo

El ciclo fundamental en Git es:

```
Editar archivos → git add → git commit → (git push)
```


````python
# Diagrama: El flujo básico de trabajo
display(mermaid('''
sequenceDiagram
    participant WD as 📁 Working Directory
    participant SA as 📋 Staging Area
    participant LR as 🗄️ Local Repo
    participant GH as ☁️ GitHub

    WD->>WD: Crear/Editar archivos
    WD->>SA: git add <archivo>
    Note over SA: Archivos listos<br/>para commit
    SA->>LR: git commit -m "mensaje"
    Note over LR: Snapshot<br/>permanente
    LR->>GH: git push origin main
    GH->>LR: git pull
    LR->>WD: git checkout / git restore
'''))
````

### `git status` — Ver el estado del repositorio

Muestra qué archivos han cambiado y en qué estado están.

```bash
git status           # Estado completo
git status -s        # Formato corto (s = short)
git status -sb       # Corto + nombre de rama actual
```

**Indicadores en formato corto:**
| Indicador | Significado |
|-----------|-------------|
| `??` | Untracked (no rastreado) |
| `A ` | Added (añadido al staging) |
| `M ` | Modified (modificado y en staging) |
| ` M` | Modified (modificado, NO en staging) |
| `MM` | Modificado en staging Y también después |
| `D ` | Deleted (eliminado) |
| `R ` | Renamed (renombrado) |


````python
%%bash
# ── Ejercicio 4.1: Explorar git status ─────────────────────────────────────
cd /tmp/git-demo

# Crear algunos archivos
echo "# Mi Proyecto" > README.md
echo "print('Hola mundo')" > main.py
mkdir -p src
echo "# Módulo principal" > src/app.py

echo "─── Estado del repositorio (formato largo) ───"
git status

echo ""
echo "─── Estado en formato corto ───"
git status -s
````

### `.gitignore` — Ignorar archivos

El archivo `.gitignore` lista patrones de archivos que Git debe ignorar.

```bash
# Ignorar por extensión
*.log
*.pyc
*.class

# Ignorar directorios
node_modules/
__pycache__/
.venv/
dist/
build/

# Ignorar archivos específicos
.env
.DS_Store
Thumbs.db

# Ignorar todo EXCEPTO un archivo (!)
*.txt
!importante.txt

# Patrones con comodines
logs/**/*.log
temp-*
```


````python
%%bash
# ── Ejercicio 4.2: Crear .gitignore ────────────────────────────────────────
cd /tmp/git-demo

cat > .gitignore << 'EOF'
# Python
__pycache__/
*.py[cod]
*.pyo
.venv/
venv/
*.egg-info/

# Entornos
.env
.env.local

# Sistema
.DS_Store
Thumbs.db

# Logs
*.log
logs/

# IDEs
.vscode/settings.json
.idea/
*.swp
EOF

echo "─── Contenido del .gitignore ───"
cat .gitignore

echo ""
# Crear archivos que deberían ignorarse
echo "mi secreto" > .env
mkdir -p logs
echo "error 404" > logs/app.log

echo "─── git status ahora (los archivos ignorados no aparecen) ───"
git status
````

### `git add` — Agregar archivos al Staging Area

```bash
git add archivo.txt          # Añadir un archivo específico
git add src/                 # Añadir todos los archivos en un directorio
git add *.py                 # Añadir archivos por patrón
git add .                    # Añadir TODOS los cambios del directorio actual
git add -A                   # Añadir TODO (incluyendo eliminaciones)
git add -p                   # Interactivo: elegir qué partes añadir (hunks)
git add -u                   # Solo archivos ya rastreados (tracked)
```

> ⚠️ **`git add .`** añade todo en el directorio actual. Usa `.gitignore` para excluir archivos no deseados.


````python
%%bash
# ── Ejercicio 4.3: Staging ─────────────────────────────────────────────────
cd /tmp/git-demo

echo "─── Añadir archivos al staging ───"
git add README.md
git add main.py src/ .gitignore

echo "─── Estado después del add ───"
git status

echo ""
echo "─── Ver qué hay exactamente en el staging ───"
git diff --cached --stat
````

### `git commit` — Confirmar los cambios

Un **commit** es un snapshot permanente del proyecto en un momento dado.  
Cada commit tiene un ID único (hash SHA-1/SHA-256).

```bash
git commit -m "mensaje del commit"         # Commit con mensaje inline
git commit                                  # Abre el editor para escribir el mensaje
git commit -am "mensaje"                    # Add + commit (solo tracked files)
git commit --amend                          # Modificar el ÚLTIMO commit
git commit --amend -m "nuevo mensaje"       # Cambiar solo el mensaje del último commit
git commit --allow-empty -m "vacío"        # Commit sin cambios (útil para CI)
```

### Anatomía de un buen mensaje de commit

```
<tipo>(<ámbito>): <resumen> (max 72 caracteres)

[cuerpo opcional - explica el QUÉ y el POR QUÉ, no el CÓMO]

[pie opcional: referencias a issues]
Closes #123
```

**Tipos según Conventional Commits:**
`feat` · `fix` · `docs` · `style` · `refactor` · `test` · `chore` · `ci` · `perf`


````python
%%bash
# ── Ejercicio 4.4: Primer commit ────────────────────────────────────────────
cd /tmp/git-demo

# Configurar identidad local (para el ejercicio)
git config user.name  "Demo User"
git config user.email "demo@tutorial.com"

# Hacer el primer commit
git commit -m "feat: agregar estructura inicial del proyecto

- README.md con descripción del proyecto
- main.py con punto de entrada
- src/app.py con módulo principal
- .gitignore para Python"

echo ""
echo "─── ¡Commit creado! ───"
git log --oneline
````

````python
%%bash
# ── Ejercicio 4.5: Hacer más commits ───────────────────────────────────────
cd /tmp/git-demo

# Segundo commit
echo '# Módulo de configuración.' > src/config.py
echo 'DEBUG = True' >> src/config.py
echo 'VERSION = "1.0.0"' >> src/config.py
git add src/config.py
git commit -m "feat: agregar módulo de configuración"

# Tercer commit
echo '# Changelog' > CHANGELOG.md
echo '## v1.0.0 - Versión inicial' >> CHANGELOG.md
git add CHANGELOG.md
git commit -m "docs: agregar CHANGELOG"

echo "─── Historial de commits ───"
git log --oneline
````

## 5. Historial y Diferencias

```bash

```

### `git diff` — Ver diferencias entre versiones

```bash
git diff                      # Cambios NO en staging (working vs staging)
git diff --staged             # Cambios EN staging (staging vs último commit)
git diff HEAD                 # Todos los cambios (working vs último commit)
git diff HEAD~1               # Comparar con el penúltimo commit
git diff <hash1> <hash2>      # Diferencia entre dos commits
git diff main..feature        # Diferencia entre dos ramas
git diff --stat               # Solo estadísticas (sin el diff completo)
git diff --word-diff          # Resaltar palabras cambiadas
git diff archivo.txt          # Solo un archivo
```

**Cómo leer un diff:**
```diff
--- a/archivo.txt          ← versión anterior
+++ b/archivo.txt          ← versión nueva
@@ -1,3 +1,4 @@           ← contexto (líneas anteriores / nuevas)
 línea sin cambio
-línea eliminada           ← rojo, prefijo -
+línea agregada            ← verde, prefijo +
 otra línea sin cambio
```


````python
%%bash
# ── Ejercicio 5.3: Usar git diff ───────────────────────────────────────────
cd /tmp/git-demo

# Hacer un cambio sin añadir al staging
echo "" >> README.md
echo "## Descripción" >> README.md
echo "Este proyecto es un ejemplo de tutorial de Git." >> README.md

echo "─── Cambios en working directory (aún no en staging) ───"
git diff

echo ""
echo "─── Añadir al staging y ver la diferencia ───"
git add README.md
git diff --staged
````

## 6. Ramas (Branches)

Una **rama** es un puntero ligero a un commit específico.  
Crear ramas en Git es casi instantáneo y no copia ningún archivo.

### ¿Para qué sirven las ramas?
- Desarrollar funcionalidades sin afectar el código estable
- Trabajar en paralelo con otros desarrolladores
- Experimentar sin riesgo
- Mantener versiones de producción separadas


````python
# Diagrama: Concepto de ramas
display(mermaid('''
gitGraph
   commit id: "C1: Initial commit"
   commit id: "C2: Add README"
   branch feature/login
   checkout feature/login
   commit id: "C3: Add login form"
   commit id: "C4: Add validation"
   checkout main
   commit id: "C5: Fix typo"
   branch hotfix/404
   checkout hotfix/404
   commit id: "C6: Fix 404 error"
   checkout main
   merge hotfix/404 id: "C7: Merge hotfix"
   merge feature/login id: "C8: Merge feature"
'''))
````

### Comandos de Branches

```bash
# ─── Ver ramas ─────────────────────────────────────────────────
git branch               # Listar ramas locales
git branch -a            # Todas (locales + remotas)
git branch -r            # Solo remotas
git branch -v            # Con el último commit de cada rama
git branch --merged      # Ramas ya fusionadas en la actual
git branch --no-merged   # Ramas aún no fusionadas

# ─── Crear ramas ────────────────────────────────────────────────
git branch nombre-rama              # Crear rama (sin cambiar a ella)
git switch -c nombre-rama           # Crear Y cambiar a ella ⭐ (moderno)
git checkout -b nombre-rama         # Idem (forma antigua)
git branch nombre-rama <hash>       # Crear rama desde un commit específico

# ─── Cambiar de rama ────────────────────────────────────────────
git switch nombre-rama              # ⭐ Moderno (Git ≥ 2.23)
git checkout nombre-rama            # Forma clásica

# ─── Renombrar y eliminar ────────────────────────────────────────
git branch -m nombre-viejo nombre-nuevo   # Renombrar
git branch -d nombre-rama                  # Eliminar (segura, si ya fusionada)
git branch -D nombre-rama                  # Eliminar forzado
git push origin --delete nombre-rama       # Eliminar rama remota
```

## 7. Fusiones (Merges)

**Merge** integra los cambios de una rama en otra.  
Git tiene varios tipos de merge según el historial de commits.


````python
# Diagrama: Fast-Forward Merge
display(mermaid('''
gitGraph
   commit id: "A"
   commit id: "B"
   branch feature
   commit id: "C"
   commit id: "D"
   checkout main
   merge feature id: "→ Fast-Forward (mueve el puntero)"
'''))
````

````python
# Diagrama: 3-Way Merge (merge commit)
display(mermaid('''
gitGraph
   commit id: "A"
   commit id: "B"
   branch feature
   commit id: "C"
   commit id: "D"
   checkout main
   commit id: "E"
   merge feature id: "M: Merge commit"
'''))
````

### Tipos de Merge

| Tipo | Cuándo ocurre | Resultado |
|------|--------------|-----------|
| **Fast-Forward** | `main` no tiene nuevos commits | Mueve el puntero, sin merge commit |
| **3-Way Merge** | Ambas ramas tienen commits nuevos | Crea un merge commit |
| **Squash Merge** | `--squash` | Aplana todos los commits en uno |
| **Rebase Merge** | `git rebase` | Historial lineal, reescribe commits |

```bash
git merge feature-rama             # Merge estándar
git merge --no-ff feature-rama     # Forzar merge commit (recomendado)
git merge --squash feature-rama    # Aplanar commits
git merge --abort                  # Cancelar un merge con conflictos
git merge --continue               # Continuar tras resolver conflictos
```


````python
%%bash
# ── Ejercicio 7.1: Fast-Forward Merge ──────────────────────────────────────
cd /tmp/git-demo

# Crear rama hotfix desde main
git switch -c hotfix/typo-readme

# Corregir algo en README
sed -i '' 's/ejemplo/DEMO/' README.md 2>/dev/null || sed -i 's/ejemplo/DEMO/' README.md
git add README.md
git commit -m "fix: corregir typo en README"

# Volver a main y fusionar (fast-forward porque main no cambió)
git switch main
git merge hotfix/typo-readme

echo "─── Historial tras el merge ───"
git log --oneline --graph --all

echo ""
echo "─── Eliminar la rama ya fusionada ───"
git branch -d hotfix/typo-readme
git log --oneline --graph --all
````

### Resolución de Conflictos

Un **conflicto** ocurre cuando dos ramas cambiaron la misma línea del mismo archivo.

```
<<<<<<< HEAD                    ← Cambios de tu rama actual
contenido de tu rama
=======                         ← Separador
contenido de la otra rama
>>>>>>> feature/otra-rama       ← Cambios de la rama a fusionar
```

**Pasos para resolver un conflicto:**
1. `git merge rama-con-conflicto` → Git marca los conflictos
2. Abrir los archivos conflictivos → Editar y elegir qué conservar
3. Eliminar los marcadores `<<<<<<<`, `=======`, `>>>>>>>`
4. `git add archivo-resuelto`
5. `git commit`  ← Git crea el merge commit automáticamente


````python
%%bash
# ── Ejercicio 7.3: Simular y resolver un conflicto ─────────────────────────
cd /tmp/git-demo

# Crear conflicto deliberado
git switch -c rama-a
echo "Versión A: función mejorada" > src/config.py
git add src/config.py
git commit -m "feat: versión A de config"

git switch main
echo "Versión B: función optimizada" > src/config.py
git add src/config.py
git commit -m "feat: versión B de config"

echo "─── Intentar fusionar (generará conflicto) ───"
git merge rama-a || true

echo ""
echo "─── Ver el conflicto ───"
cat src/config.py

echo ""
echo "─── Resolver el conflicto ───"
echo '# Configuración combinada.' > src/config.py
echo 'DEBUG = True' >> src/config.py
echo 'VERSION = "1.0.0"' >> src/config.py
echo '# Versión unificada de A y B' >> src/config.py

git add src/config.py
git commit -m "fix: resolver conflicto de merge en config.py"

echo ""
echo "─── Historial tras resolver conflicto ───"
git log --oneline --graph

# Limpiar
git branch -D rama-a
````

```bash

git push origin main               # Push de la rama main
git push origin feature/nueva      # Push de una rama específica
git push -u origin main            # Push + configurar upstream (-u)
git push --all                     # Push de todas las ramas
git push origin :rama-remota       # Eliminar rama remota (forma antigua)
git push origin --delete rama      # Eliminar rama remota (forma nueva)
git push --tags                    # Push de todos los tags

# ─── Fetch vs Pull ──────────────────────────────────────────────
# git fetch = solo descarga, no modifica tu rama
# git pull  = fetch + merge (puede crear conflictos)
# RECOMENDACIÓN: usa git fetch + git merge manualmente para más control
```

## 11. Deshacer Cambios

Git ofrece varias formas de deshacer cambios según en qué estado estén.


````python
# Diagrama: Árbol de decisión para deshacer
display(mermaid('''
flowchart TD
    Q{¿Qué quieres deshacer?}
    
    Q -->|Cambios no añadidos al staging| R1[git restore archivo.txt]
    Q -->|Cambios en el staging area| R2[git restore --staged archivo.txt]
    Q -->|El último commit, mantener cambios| R3[git reset --soft HEAD~1]
    Q -->|El último commit, descartar cambios| R4[git reset --hard HEAD~1]
    Q -->|Un commit antiguo, historial seguro| R5[git revert <hash>]
    Q -->|Archivos no rastreados| R6[git clean -fd]
    Q -->|Todo lo anterior | R7[⚠️ Cuidado con --hard y clean -fd]

    style R1 fill:#238636,color:white
    style R2 fill:#1f6feb,color:white
    style R3 fill:#9e6a03,color:white
    style R4 fill:#da3633,color:white
    style R5 fill:#6e40c9,color:white
    style R6 fill:#da3633,color:white
    style R7 fill:#30363d,color:#cdd9e5
'''))
````

### `git restore` — Descartar cambios (Git ≥ 2.23)

```bash
git restore archivo.txt              # Descartar cambios en working dir
git restore .                        # Descartar TODOS los cambios
git restore --staged archivo.txt     # Quitar del staging (mantener cambios)
git restore --staged .               # Quitar todo del staging
git restore --source=HEAD~2 archivo  # Restaurar versión de hace 2 commits
```

### `git reset` — Mover el puntero HEAD

```bash
git reset --soft HEAD~1    # Deshacer commit, MANTENER cambios en staging
git reset --mixed HEAD~1   # Deshacer commit, MANTENER cambios en working (default)
git reset --hard HEAD~1    # ⚠️ Deshacer commit, ELIMINAR todos los cambios
git reset <hash>           # Volver a un commit específico
```

> ⚠️ **ADVERTENCIA:** `git reset --hard` es DESTRUCTIVO. No usar en commits ya publicados.

### `git revert` — Deshacer de forma segura

```bash
git revert HEAD               # Crear commit que deshace el último commit
git revert <hash>             # Deshacer un commit específico
git revert HEAD~3..HEAD       # Deshacer los últimos 3 commits
git revert --no-commit HEAD   # Deshacer sin hacer commit automático
```

> ✅ **`git revert` es seguro** porque no reescribe el historial.  
> Úsalo cuando los commits ya están en un repositorio compartido (push).

### `git clean` — Eliminar archivos no rastreados

```bash
git clean -n               # Ver qué se eliminaría (dry-run)
git clean -f               # Eliminar archivos no rastreados
git clean -fd              # Eliminar archivos Y directorios
git clean -fdx             # Eliminar también archivos ignorados (.gitignore)
```

### `git rebase` — Reescribir el historial

**Rebase** mueve o combina commits de una rama encima de otra,  
creando un historial lineal más limpio.

```
Antes del rebase:

A -- B -- C  (main)
          \
           D -- E  (feature)

Después de: git rebase main (desde feature):

               \

```


````python
# Diagrama: git rebase
display(mermaid('''
gitGraph
   commit id: "A"
   commit id: "B"
   commit id: "C"
   branch feature
   commit id: "D (antes)"
   commit id: "E (antes)"
   checkout main
   commit id: "F"
   checkout feature
   commit id: "D_primo: D rebased" type: HIGHLIGHT
   commit id: "E_primo: E rebased" type: HIGHLIGHT
'''))
````

```bash
# ─── Rebase básico ─────────────────────────────────────────────
git rebase main               # Rebase de la rama actual sobre main
git rebase --onto main A B    # Rebase avanzado (mover commits de A a B encima de main)

git rebase --continue         # Continuar tras resolver conflictos
git rebase --skip             # Saltar el commit conflictivo

# ─── Rebase interactivo (el más poderoso) ──────────────────────
git rebase -i HEAD~3          # Editar los últimos 3 commits interactivamente
git rebase -i <hash>          # Editar desde ese commit en adelante
```

**Opciones en rebase interactivo:**
| Comando | Acción |
|---------|--------|
| `pick` | Conservar el commit tal cual |
| `reword` | Conservar pero cambiar el mensaje |
| `edit` | Pausar para modificar el commit |
| `squash` | Combinar con el commit anterior |
| `fixup` | Como squash pero descarta el mensaje |
| `drop` | Eliminar el commit |
| `exec` | Ejecutar un comando shell |

> ⚠️ **Regla de oro:** Nunca hagas rebase de commits que ya están en un repositorio compartido (remote).

### `git tag` — Etiquetar versiones

Los **tags** marcan puntos específicos del historial, típicamente para versiones de lanzamiento.

```bash
# ─── Tags ligeros (solo un puntero) ────────────────────────────
git tag v1.0.0                      # Crear tag en HEAD
git tag v1.0.0 <hash>               # Crear tag en un commit específico

# ─── Tags anotados (recomendados para releases) ─────────────────
git tag -a v1.0.0 -m "Release 1.0.0"      # Con mensaje
git tag -a v1.0.0 <hash> -m "mensaje"     # En un commit específico

# ─── Ver tags ───────────────────────────────────────────────────
git tag                             # Listar todos los tags
git tag -l "v1.*"                   # Filtrar por patrón
git show v1.0.0                     # Detalles del tag

# ─── Compartir tags ─────────────────────────────────────────────
git push origin v1.0.0              # Push de un tag específico
git push origin --tags              # Push de todos los tags

# ─── Eliminar tags ──────────────────────────────────────────────
git tag -d v1.0.0                   # Eliminar tag local
git push origin --delete v1.0.0    # Eliminar tag remoto
```


````python
%%bash
# ── Ejercicio 12.3: git tag ─────────────────────────────────────────────────
cd /tmp/git-demo

# Crear un tag anotado para la versión actual
git tag -a v1.0.0 -m "Primera versión estable del proyecto"

# Crear otro tag para un commit anterior
git tag -a v0.1.0 HEAD~3 -m "Versión inicial (alpha)"

echo "─── Tags disponibles ───"
git tag -n   # -n muestra la primera línea del mensaje

echo ""
echo "─── Detalles del tag v1.0.0 ───"
git show v1.0.0 --stat

echo ""
echo "─── Ver tags en el historial ───"
git log --oneline --decorate | head -8
````

### `git bisect` — Encontrar el commit que introdujo un bug

`git bisect` usa búsqueda binaria para encontrar el commit culpable de un bug.

```bash
git bisect start            # Iniciar sesión de bisect
git bisect bad              # El commit actual TIENE el bug
git bisect good v1.0.0      # Este commit NO tiene el bug
# Git hace checkout automáticamente al commit del medio
git bisect good             # Si este commit no tiene bug
git bisect bad              # Si este commit tiene bug
# ... repetir hasta encontrar el commit culpable ...
git bisect reset            # Terminar y volver al HEAD original

# Automático con un script de prueba:
git bisect start
git bisect bad HEAD
git bisect good v1.0.0
git bisect run ./test.sh   # 0 = bueno, !0 = malo
```


### `git blame` — Ver quién modificó cada línea

```bash
git blame archivo.txt                    # Ver todos los autores por línea
git blame -L 10,20 archivo.txt           # Solo líneas 10-20
git blame -w archivo.txt                 # Ignorar cambios de espacios
git blame --since="3 weeks ago" archivo  # Desde una fecha

# Formato de salida:
# <hash> (<autor> <fecha> <línea>) <contenido>
```


````python
%%bash
# ── Ejercicio 12.4: git blame ───────────────────────────────────────────────
cd /tmp/git-demo

echo "─── ¿Quién escribió cada línea de main.py? ───"
git blame main.py

echo ""
echo "─── Solo primeras 5 líneas de README.md ───"
git blame -L 1,5 README.md
````

### `git reflog` — El historial de todo

`reflog` registra TODOS los movimientos de HEAD, incluso los que no están en `git log`.  
Es la **red de seguridad** de Git: permite recuperar commits "perdidos".

```bash
git reflog                      # Ver historial completo de HEAD
git reflog show main            # Historial de una rama específica
git reflog expire --expire=90.days.ago --all   # Limpiar entradas viejas

# Recuperar commits "perdidos" tras un git reset --hard:
git reflog                     # Buscar el hash del commit perdido
git reset --hard <hash>        # Restaurar a ese commit
# O crear una rama desde el commit perdido:
git branch recuperado <hash>
```


````python
%%bash
# ── Ejercicio 12.5: git reflog ─────────────────────────────────────────────
cd /tmp/git-demo

echo "─── Historial completo de acciones (reflog) ───"
git reflog | head -20
````

## 13. Flujos de Trabajo (Workflows)

### Git Flow — Para proyectos con releases planificados


````python
# Diagrama: Git Flow
display(mermaid('''
gitGraph
   commit id: "Inicio"
   branch develop
   checkout develop
   commit id: "Setup inicial"
   branch feature/login
   checkout feature/login
   commit id: "Login form"
   commit id: "Login validation"
   checkout develop
   merge feature/login id: "Merge login"
   branch feature/dashboard
   checkout feature/dashboard
   commit id: "Dashboard UI"
   checkout develop
   merge feature/dashboard id: "Merge dashboard"
   branch release/1.0
   checkout release/1.0
   commit id: "Version bump"
   commit id: "Fix last bugs"
   checkout main
   merge release/1.0 id: "v1.0.0" tag: "v1.0.0"
   checkout develop
   merge release/1.0 id: "Sync develop"
'''))
````

**Ramas en Git Flow:**

| Rama | Propósito | Vida |
|------|-----------|------|
| `main` | Código en producción | Permanente |
| `develop` | Integración de features | Permanente |
| `feature/*` | Nuevas funcionalidades | Temporal |
| `release/*` | Preparar un release | Temporal |
| `hotfix/*` | Correcciones urgentes en producción | Temporal |

```bash
# Instalar git-flow (extensión)
brew install git-flow-avh

# Inicializar en un repositorio
git flow init

# Comandos de git-flow
git flow feature start mi-feature
git flow feature finish mi-feature
git flow release start 1.0.0
git flow release finish 1.0.0
git flow hotfix start fix-urgente
git flow hotfix finish fix-urgente
```

### GitHub Flow — Para despliegue continuo (más simple)


````python
# Diagrama: GitHub Flow
display(mermaid('''
flowchart LR
    M[main\n🏭 Producción] -->|1. Crear rama| F[feature/nombre]
    F -->|2. Commits| F
    F -->|3. Push + PR| PR{Pull Request\n👀 Code Review}
    PR -->|4. Aprobado| M2[main]
    PR -->|Cambios pedidos| F
    M2 -->|5. Deploy automático| D[🚀 Producción]

    style M fill:#238636,color:white
    style M2 fill:#238636,color:white
    style PR fill:#388bfd,color:white
    style D fill:#9e6a03,color:white
    style F fill:#6e40c9,color:white
'''))
````

**GitHub Flow (reglas simples):**
1. `main` siempre está desplegable en producción
2. Crear ramas descriptivas para cada tarea: `feature/añadir-búsqueda`
3. Hacer commits frecuentes con mensajes claros
4. Abrir un Pull Request pronto (permite discusión temprana)
5. Revisión de código antes de hacer merge
6. Deploy inmediatamente tras merge a main

### Trunk-Based Development — Para equipos avanzados con CI/CD

```
✅ Una sola rama principal (main/trunk)
✅ Feature flags para código incompleto
✅ Commits frecuentes (varias veces al día)
✅ CI/CD muy robusto
✅ Ramas muy cortas (máx 1-2 días)
```

### ¿Qué flujo elegir?

| Criterio | Git Flow | GitHub Flow | Trunk-Based |
|----------|----------|-------------|-------------|
| Tamaño del equipo | Mediano-grande | Cualquiera | Grande + maduro |
| Frecuencia de releases | Planificados | Continuo | Continuo |
| CI/CD | Opcional | Recomendado | Obligatorio |
| Complejidad | Alta | Baja | Media |
| Mejor para | Apps con versiones | Startups / web | Google / Netflix |

## 14. Buenas Prácticas

### Conventional Commits

Estándar de la industria para mensajes de commit estructurados.  
Habilita generación automática de CHANGELOG y versionado semántico.

```
<tipo>[ámbito opcional]: <descripción>

[cuerpo opcional]

[pie(s) opcional(es)]
```

**Tipos:**
| Tipo | Descripción | SemVer |
|------|-------------|--------|
| `feat` | Nueva funcionalidad | MINOR |
| `fix` | Corrección de bug | PATCH |
| `docs` | Solo documentación | — |
| `style` | Formato, sin lógica | — |
| `refactor` | Refactorización | — |
| `test` | Añadir/arreglar tests | — |
| `chore` | Tareas de mantenimiento | — |
| `ci` | Cambios en CI/CD | — |
| `perf` | Mejoras de rendimiento | PATCH |
| `BREAKING CHANGE` | Cambio incompatible | MAJOR |

**Ejemplos:**
```
feat(auth): agregar autenticación con Google OAuth

fix(api): corregir error 500 en endpoint /users

docs(readme): actualizar instrucciones de instalación

feat!: cambiar la API de autenticación
BREAKING CHANGE: el parámetro 'token' ahora se llama 'access_token'
```

### Reglas de los commits atómicos

1. **Un solo propósito por commit** — Si necesitas "y" para describir el commit, divide en dos
2. **Commits pequeños y frecuentes** — Más fáciles de revisar y revertir
3. **El proyecto debe funcionar en cada commit** — Nunca commit de código roto
4. **Mensaje claro y en imperativo** — "Add login page" no "Added login page"

## 15. Cheat Sheet — Referencia Rápida Completa

### 🔧 Configuración
```bash
git config --global user.name "Nombre"
git config --global user.email "email@mail.com"
git config --global init.defaultBranch main
git config --global alias.lg "log --oneline --graph --decorate --all"
git config --list
```

### 📁 Repositorios
```bash
git init                          # Nuevo repositorio
git clone <url>                   # Clonar repositorio
git clone <url> --depth 1         # Clone superficial (rápido)
```

### 📊 Estado e Información
```bash
git status                        # Estado actual
git status -s                     # Estado corto
git log --oneline --graph --all   # Historial gráfico
git diff                          # Cambios sin staging
git diff --staged                 # Cambios en staging
git show <hash>                   # Detalles de un commit
git blame archivo.txt             # Quién escribió cada línea
```

### ➕ Staging y Commits
```bash
git add archivo.txt               # Añadir archivo
git add .                         # Añadir todo
git add -p                        # Añadir interactivo (hunks)
git commit -m "mensaje"           # Commit con mensaje
git commit --amend                # Modificar último commit
git commit -am "mensaje"          # Add tracked + commit
```

### 🌿 Ramas
```bash
git branch                        # Listar ramas
git branch -a                     # Listar todas (locales + remotas)
git switch -c nueva-rama          # Crear y cambiar
git switch main                   # Cambiar de rama
git branch -d rama                # Eliminar rama (seguro)
git branch -D rama                # Eliminar forzado
git merge rama                    # Fusionar rama
git merge --no-ff rama            # Fusionar con merge commit
```

### ☁️ Remotos
```bash
git remote -v                     # Ver remotos
git remote add origin <url>       # Añadir remoto
git fetch origin                  # Descargar sin fusionar
git pull                          # Descargar y fusionar
git push origin main              # Subir cambios
git push -u origin main           # Subir + configurar upstream
git push origin --delete rama     # Eliminar rama remota
```

### 💼 Stash
```bash
git stash                         # Guardar cambios temporalmente
git stash push -m "descripción"   # Guardar con nombre
git stash list                    # Ver stashes
git stash pop                     # Recuperar último stash
git stash drop stash@{0}          # Eliminar stash
```

### ↩️ Deshacer
```bash
git restore archivo.txt           # Descartar cambios
git restore --staged archivo.txt  # Quitar del staging
git reset --soft HEAD~1           # Deshacer commit (mantener staging)
git reset --hard HEAD~1           # ⚠️ Deshacer commit y cambios
git revert HEAD                   # Revertir commit (seguro)
git clean -fd                     # Eliminar untracked
```

### 🔥 Avanzados
```bash
git rebase main                   # Rebase sobre main
git rebase -i HEAD~3              # Rebase interactivo
git cherry-pick <hash>            # Aplicar un commit específico
git tag -a v1.0.0 -m "Release"   # Crear tag anotado
git reflog                        # Historial de todo HEAD
git bisect start/good/bad/reset   # Buscar commit con bug
```
