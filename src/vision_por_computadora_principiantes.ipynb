## 1. CONCEPTOS FUNDAMENTALES

### ¿Cómo representa una computadora una imagen?

Una imagen digital se compone de:
- **Píxeles**: Son los "ladrillos" de una imagen. Cada píxel tiene una posición (x, y) y un valor de color.
- **Resolución**: Se mide en píxeles (ej: 1920x1080)
- **Canales de color**: 
  - **Escala de grises**: 1 canal (0-255, donde 0 es negro y 255 es blanco)
  - **RGB**: 3 canales (Rojo, Verde, Azul)
  - **RGBA**: 4 canales (RGB + Alfa para transparencia)

### Matriz de píxeles
```
Una imagen en escala de grises es una matriz 2D:
┌─────────────┐
│ [0  255  10] │
│ [50  100 200] │
│ [255 0  128] │
└─────────────┘

Una imagen RGB es una matriz 3D:
┌────────────────────────┐
│ [[R,G,B] [R,G,B] ...] │
│ [[R,G,B] [R,G,B] ...] │
│ ...                     │
└────────────────────────┘
```

## 3. CARGAR Y VISUALIZAR IMÁGENES

### Formas de cargar una imagen:
1. **Desde archivo local**
2. **Desde URL (internet)**
3. **Crear una imagen nueva**


````python
# Crear una imagen de prueba simple
# Imagen RGB con colores gradientes

imagen = np.zeros((300, 400, 3), dtype=np.uint8)

# Llenar con gradiente rojo
for i in range(300):
    imagen[i, :, 0] = int((i / 300) * 255)  # Rojo
    imagen[i, :, 1] = int((i / 300) * 128)  # Verde
    imagen[i, :, 2] = 100  # Azul constante

# Mostrar información de la imagen
print(f"Forma (shape): {imagen.shape}")
print(f"Altura: {imagen.shape[0]} píxeles")
print(f"Ancho: {imagen.shape[1]} píxeles")
print(f"Canales: {imagen.shape[2]} (RGB)")
print(f"Tipo de dato: {imagen.dtype}")
print(f"Valores mín-máx: {imagen.min()}-{imagen.max()}")
````

````python
# Visualizar la imagen
plt.figure(figsize=(10, 6))
plt.imshow(imagen)
plt.title('Imagen con Gradiente', fontsize=14, fontweight='bold')
plt.axis('off')  # Ocultar ejes
plt.tight_layout()
plt.show()

print("✅ Imagen visualizada correctamente")
````

## 4. DESCARGAR UNA IMAGEN DE EJEMPLO

Vamos a descargar una imagen real de internet para trabajar con ella.


````python
# Descargar imagen de ejemplo (usando una imagen de prueba conocida)
url = 'https://upload.wikimedia.org/wikipedia/commons/thumb/a/a7/Camponotus_flavomarginatus_ant.jpg/320px-Camponotus_flavomarginatus_ant.jpg'

try:
    urllib.request.urlretrieve(url, 'imagen_prueba.jpg')
    print("✅ Imagen descargada exitosamente")
    
    # Cargar imagen con OpenCV (nota: OpenCV carga en BGR, no RGB)
    img_bgr = cv2.imread('imagen_prueba.jpg')
    
    # Convertir BGR a RGB para matplotlib
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
    
    print(f"Imagen cargada: {img_rgb.shape}")
    
    # Visualizar
    plt.figure(figsize=(10, 6))
    plt.imshow(img_rgb)
    plt.title('Imagen de Prueba', fontsize=14, fontweight='bold')
    plt.axis('off')
    plt.tight_layout()
    plt.show()
except Exception as e:
    print(f"Error: {e}")
    print("Continuaremos con imágenes sintetizadas")
````

## 5. OPERACIONES BÁSICAS CON IMÁGENES

### 5.1 Cambiar tamaño (Resize)


````python
# Crear una imagen de ejemplo
imagen_original = np.random.randint(0, 256, (600, 800, 3), dtype=np.uint8)

# Cambiar tamaño
imagen_pequeña = cv2.resize(imagen_original, (400, 300))
imagen_grande = cv2.resize(imagen_original, (1200, 900))

print(f"Imagen original: {imagen_original.shape}")
print(f"Imagen pequeña: {imagen_pequeña.shape}")
print(f"Imagen grande: {imagen_grande.shape}")

# Visualizar
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].imshow(imagen_original)
axes[0].set_title('Original (800x600)', fontweight='bold')
axes[0].axis('off')

axes[1].imshow(imagen_pequeña)
axes[1].set_title('Reducida (400x300)', fontweight='bold')
axes[1].axis('off')

axes[2].imshow(imagen_grande)
axes[2].set_title('Ampliada (1200x900)', fontweight='bold')
axes[2].axis('off')

plt.tight_layout()
plt.show()
````

### 5.2 Convertir a Escala de Grises


````python
# Crear imagen RGB de ejemplo
img_rgb = np.zeros((300, 400, 3), dtype=np.uint8)

# Llenar con colores
img_rgb[0:100, :] = [255, 0, 0]      # Rojo
img_rgb[100:200, :] = [0, 255, 0]    # Verde
img_rgb[200:300, :] = [0, 0, 255]    # Azul

# Convertir a escala de grises
img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)

print(f"Imagen RGB shape: {img_rgb.shape}")
print(f"Imagen Gris shape: {img_gray.shape}")

# Visualizar
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].imshow(img_rgb)
axes[0].set_title('Imagen RGB (3 canales)', fontweight='bold')
axes[0].axis('off')

axes[1].imshow(img_gray, cmap='gray')
axes[1].set_title('Imagen en Escala de Grises (1 canal)', fontweight='bold')
axes[1].axis('off')

plt.tight_layout()
plt.show()
````

### 5.3 Rotar Imagen


````python
# Crear imagen con patrón para ver la rotación
imagen = np.ones((300, 400, 3), dtype=np.uint8) * 200

# Agregar texto para ver la rotación
cv2.putText(imagen, 'ORIGINAL', (50, 150), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)

# Obtener centro de rotación
centro = (imagen.shape[1] // 2, imagen.shape[0] // 2)

# Crear matriz de rotación (45 grados)
matriz_rotacion = cv2.getRotationMatrix2D(centro, 45, 1.0)
imagen_rotada = cv2.warpAffine(imagen, matriz_rotacion, (imagen.shape[1], imagen.shape[0]))

# Visualizar
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].imshow(cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB))
axes[0].set_title('Original', fontweight='bold')
axes[0].axis('off')

axes[1].imshow(cv2.cvtColor(imagen_rotada, cv2.COLOR_BGR2RGB))
axes[1].set_title('Rotada 45°', fontweight='bold')
axes[1].axis('off')

plt.tight_layout()
plt.show()
````

## 6. FILTROS Y TRANSFORMACIONES

### 6.1 Blur (Desenfoque)


````python
# Crear imagen con ruido
np.random.seed(42)
imagen_ruidosa = np.random.randint(0, 256, (300, 400, 3), dtype=np.uint8)

# Aplicar diferentes niveles de blur
blur_3 = cv2.blur(imagen_ruidosa, (3, 3))
blur_7 = cv2.blur(imagen_ruidosa, (7, 7))
blur_15 = cv2.blur(imagen_ruidosa, (15, 15))

# Visualizar
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

axes[0, 0].imshow(imagen_ruidosa)
axes[0, 0].set_title('Original (con ruido)', fontweight='bold')
axes[0, 0].axis('off')

axes[0, 1].imshow(blur_3)
axes[0, 1].set_title('Blur (3x3)', fontweight='bold')
axes[0, 1].axis('off')

axes[1, 0].imshow(blur_7)
axes[1, 0].set_title('Blur (7x7)', fontweight='bold')
axes[1, 0].axis('off')

axes[1, 1].imshow(blur_15)
axes[1, 1].set_title('Blur (15x15)', fontweight='bold')
axes[1, 1].axis('off')

plt.tight_layout()
plt.show()

print("Mayor tamaño del kernel = mayor desenfoque")
````

### 6.2 Detección de Bordes con Canny


````python
# Crear imagen con formas
imagen_formas = np.ones((300, 400, 3), dtype=np.uint8) * 255

# Dibujar formas
cv2.circle(imagen_formas, (100, 100), 50, (0, 0, 255), -1)  # Círculo rojo
cv2.rectangle(imagen_formas, (250, 50), (350, 150), (0, 255, 0), -1)  # Rectángulo verde
cv2.line(imagen_formas, (50, 250), (350, 250), (255, 0, 0), 3)  # Línea azul

# Convertir a escala de grises
imagen_gray = cv2.cvtColor(imagen_formas, cv2.COLOR_BGR2GRAY)

# Aplicar Canny (detección de bordes)
bordes = cv2.Canny(imagen_gray, 50, 150)

# Visualizar
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].imshow(cv2.cvtColor(imagen_formas, cv2.COLOR_BGR2RGB))
axes[0].set_title('Imagen Original con Formas', fontweight='bold')
axes[0].axis('off')

axes[1].imshow(bordes, cmap='gray')
axes[1].set_title('Bordes Detectados (Canny)', fontweight='bold')
axes[1].axis('off')

plt.tight_layout()
plt.show()

print("Canny detecta cambios abruptos de intensidad (bordes)")
````

### 6.3 Detección de Características: Esquinas (Harris)


````python
# Crear imagen con esquinas claras
imagen_esquinas = np.ones((300, 400), dtype=np.uint8) * 255
cv2.rectangle(imagen_esquinas, (50, 50), (250, 200), 0, -1)
cv2.circle(imagen_esquinas, (150, 150), 30, 255, -1)

# Convertir a float32
imagen_float = np.float32(imagen_esquinas)

# Detectar esquinas con Harris
esquinas = cv2.cornerHarris(imagen_float, 2, 3, 0.04)

# Umbralizar para obtener esquinas más claras
imagen_con_esquinas = imagen_esquinas.copy()
imagen_con_esquinas[esquinas > 0.01 * esquinas.max()] = 100

# Visualizar
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].imshow(imagen_esquinas, cmap='gray')
axes[0].set_title('Imagen Original', fontweight='bold')
axes[0].axis('off')

axes[1].imshow(imagen_con_esquinas, cmap='gray')
axes[1].set_title('Esquinas Detectadas (Harris)', fontweight='bold')
axes[1].axis('off')

plt.tight_layout()
plt.show()

print("Harris detecta puntos de interés (esquinas)")
````

## 7. HISTOGRAMAS

Un histograma muestra la distribución de valores de píxeles en una imagen.


````python
# Crear imagen con diferentes intensidades
imagen_hist = np.zeros((300, 400), dtype=np.uint8)
imagen_hist[0:100, :] = 50       # Oscuro
imagen_hist[100:200, :] = 150    # Medio
imagen_hist[200:300, :] = 250    # Claro

# Calcular histograma
histograma = cv2.calcHist([imagen_hist], [0], None, [256], [0, 256])

# Visualizar
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].imshow(imagen_hist, cmap='gray')
axes[0].set_title('Imagen (3 niveles de intensidad)', fontweight='bold')
axes[0].axis('off')

axes[1].plot(histograma, color='black')
axes[1].set_title('Histograma', fontweight='bold')
axes[1].set_xlabel('Valor de Píxel')
axes[1].set_ylabel('Frecuencia')
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("El histograma muestra tres picos correspondientes a los tres niveles")
````

## 8. UMBRALIZACION (THRESHOLDING)

Convertir una imagen en una imagen binaria (solo blanco y negro).


````python
# Crear imagen con gradiente
imagen_gradiente = np.zeros((300, 400), dtype=np.uint8)
for i in range(400):
    imagen_gradiente[:, i] = int((i / 400) * 255)

# Aplicar diferentes umbrales
ret1, umbral_100 = cv2.threshold(imagen_gradiente, 100, 255, cv2.THRESH_BINARY)
ret2, umbral_150 = cv2.threshold(imagen_gradiente, 150, 255, cv2.THRESH_BINARY)
ret3, umbral_200 = cv2.threshold(imagen_gradiente, 200, 255, cv2.THRESH_BINARY)

# Visualizar
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

axes[0, 0].imshow(imagen_gradiente, cmap='gray')
axes[0, 0].set_title('Original (Gradiente)', fontweight='bold')
axes[0, 0].axis('off')

axes[0, 1].imshow(umbral_100, cmap='gray')
axes[0, 1].set_title('Umbral = 100', fontweight='bold')
axes[0, 1].axis('off')

axes[1, 0].imshow(umbral_150, cmap='gray')
axes[1, 0].set_title('Umbral = 150', fontweight='bold')
axes[1, 0].axis('off')

axes[1, 1].imshow(umbral_200, cmap='gray')
axes[1, 1].set_title('Umbral = 200', fontweight='bold')
axes[1, 1].axis('off')

plt.tight_layout()
plt.show()

print("Pixeles > umbral = 255 (blanco)")
print("Pixeles <= umbral = 0 (negro)")
````

## 9. EJERCICIO PRÁCTICO: PROCESAMIENTO COMPLETO

Vamos a aplicar múltiples transformaciones a una imagen.


````python
# Crear imagen de ejemplo con ruido
np.random.seed(123)
imagen_base = np.zeros((300, 400, 3), dtype=np.uint8)
cv2.circle(imagen_base, (200, 150), 80, (0, 255, 255), -1)  # Círculo amarillo
cv2.rectangle(imagen_base, (50, 50), (150, 250), (255, 0, 0), -1)  # Rectángulo azul

# Agregar ruido
ruido = np.random.randint(-50, 50, imagen_base.shape).astype(np.int16)
imagen_ruidosa = np.clip(imagen_base.astype(np.int16) + ruido, 0, 255).astype(np.uint8)

# Paso 1: Convertir a escala de grises
imagen_gray = cv2.cvtColor(imagen_ruidosa, cv2.COLOR_BGR2GRAY)

# Paso 2: Aplicar blur para reducir ruido
imagen_suavizada = cv2.blur(imagen_gray, (5, 5))

# Paso 3: Detectar bordes
bordes = cv2.Canny(imagen_suavizada, 50, 150)

# Visualizar pipeline
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

axes[0, 0].imshow(cv2.cvtColor(imagen_base, cv2.COLOR_BGR2RGB))
axes[0, 0].set_title('Original (sin ruido)', fontweight='bold')
axes[0, 0].axis('off')

axes[0, 1].imshow(cv2.cvtColor(imagen_ruidosa, cv2.COLOR_BGR2RGB))
axes[0, 1].set_title('Con ruido', fontweight='bold')
axes[0, 1].axis('off')

axes[0, 2].imshow(imagen_gray, cmap='gray')
axes[0, 2].set_title('Escala de grises', fontweight='bold')
axes[0, 2].axis('off')

axes[1, 0].imshow(imagen_suavizada, cmap='gray')
axes[1, 0].set_title('Suavizada (blur 5x5)', fontweight='bold')
axes[1, 0].axis('off')

axes[1, 1].imshow(bordes, cmap='gray')
axes[1, 1].set_title('Bordes detectados', fontweight='bold')
axes[1, 1].axis('off')

axes[1, 2].axis('off')

plt.tight_layout()
plt.show()

print("Pipeline de procesamiento completo:")
print("1. Imagen ruidosa -> 2. Escala de grises -> 3. Suavizado -> 4. Detección de bordes")
````
