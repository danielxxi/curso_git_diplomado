## 1. CONCEPTOS FUNDAMENTALES

### ¿Cómo representa una computadora una imagen?

Una imagen digital se compone de:
- **Píxeles**: Son los "ladrillos" de una imagen. Cada píxel tiene una posición (x, y) y un valor de color.
- **Resolución**: Se mide en píxeles (ej: 1920x1080)
- **Canales de color**: 
  - **Escala de grises**: 1 canal (0-255, donde 0 es negro y 255 es blanco)
  - **RGB**: 3 canales (Rojo, Verde, Azul)
  - **RGBA**: 4 canales (RGB + Alfa para transparencia)

### Matriz de píxeles
```
Una imagen en escala de grises es una matriz 2D:
┌─────────────┐
│ [0  255  10] │
│ [50  100 200] │
│ [255 0  128] │
└─────────────┘

Una imagen RGB es una matriz 3D:
┌────────────────────────┐
│ [[R,G,B] [R,G,B] ...] │
│ [[R,G,B] [R,G,B] ...] │
│ ...                     │
└────────────────────────┘
```

## 3. CARGAR Y VISUALIZAR IMÁGENES

### Formas de cargar una imagen:
1. **Desde archivo local**
2. **Desde URL (internet)**
3. **Crear una imagen nueva**


````python
# Crear una imagen de prueba simple
# Imagen RGB con colores gradientes

imagen = np.zeros((300, 400, 3), dtype=np.uint8)

# Llenar con gradiente rojo
for i in range(300):
    imagen[i, :, 0] = int((i / 300) * 255)  # Rojo
    imagen[i, :, 1] = int((i / 300) * 128)  # Verde
    imagen[i, :, 2] = 100  # Azul constante

# Mostrar información de la imagen
print(f"Forma (shape): {imagen.shape}")
print(f"Altura: {imagen.shape[0]} píxeles")
print(f"Ancho: {imagen.shape[1]} píxeles")
print(f"Canales: {imagen.shape[2]} (RGB)")
print(f"Tipo de dato: {imagen.dtype}")
print(f"Valores mín-máx: {imagen.min()}-{imagen.max()}")
````

````python
# Visualizar la imagen
plt.figure(figsize=(10, 6))
plt.imshow(imagen)
plt.title('Imagen con Gradiente', fontsize=14, fontweight='bold')
plt.axis('off')  # Ocultar ejes
plt.tight_layout()
plt.show()

print("✅ Imagen visualizada correctamente")
````

## 4. OPERACIONES BÁSICAS CON IMÁGENES

### 4.1 Cambiar tamaño (Resize)


````python
# Crear una imagen de ejemplo
imagen_original = np.random.randint(0, 256, (600, 800, 3), dtype=np.uint8)

# Cambiar tamaño
imagen_pequeña = cv2.resize(imagen_original, (400, 300))
imagen_grande = cv2.resize(imagen_original, (1200, 900))

print(f"Imagen original: {imagen_original.shape}")
print(f"Imagen pequeña: {imagen_pequeña.shape}")
print(f"Imagen grande: {imagen_grande.shape}")

# Visualizar
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].imshow(imagen_original)
axes[0].set_title('Original (800x600)', fontweight='bold')
axes[0].axis('off')

axes[1].imshow(imagen_pequeña)
axes[1].set_title('Reducida (400x300)', fontweight='bold')
axes[1].axis('off')

axes[2].imshow(imagen_grande)
axes[2].set_title('Ampliada (1200x900)', fontweight='bold')
axes[2].axis('off')

plt.tight_layout()
plt.show()
````

### 4.2 Convertir a Escala de Grises


````python
# Crear imagen RGB de ejemplo
img_rgb = np.zeros((300, 400, 3), dtype=np.uint8)

# Llenar con colores
img_rgb[0:100, :] = [255, 0, 0]      # Rojo
img_rgb[100:200, :] = [0, 255, 0]    # Verde
img_rgb[200:300, :] = [0, 0, 255]    # Azul

# Convertir a escala de grises
img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)

print(f"Imagen RGB shape: {img_rgb.shape}")
print(f"Imagen Gris shape: {img_gray.shape}")

# Visualizar
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].imshow(img_rgb)
axes[0].set_title('Imagen RGB (3 canales)', fontweight='bold')
axes[0].axis('off')

axes[1].imshow(img_gray, cmap='gray')
axes[1].set_title('Imagen en Escala de Grises (1 canal)', fontweight='bold')
axes[1].axis('off')

plt.tight_layout()
plt.show()
````

## 5. FILTROS Y TRANSFORMACIONES

### 5.1 Blur (Desenfoque)


````python
# Crear imagen con ruido
np.random.seed(42)
imagen_ruidosa = np.random.randint(0, 256, (300, 400, 3), dtype=np.uint8)

# Aplicar diferentes niveles de blur
blur_3 = cv2.blur(imagen_ruidosa, (3, 3))
blur_7 = cv2.blur(imagen_ruidosa, (7, 7))
blur_15 = cv2.blur(imagen_ruidosa, (15, 15))

# Visualizar
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

axes[0, 0].imshow(imagen_ruidosa)
axes[0, 0].set_title('Original (con ruido)', fontweight='bold')
axes[0, 0].axis('off')

axes[0, 1].imshow(blur_3)
axes[0, 1].set_title('Blur (3x3)', fontweight='bold')
axes[0, 1].axis('off')

axes[1, 0].imshow(blur_7)
axes[1, 0].set_title('Blur (7x7)', fontweight='bold')
axes[1, 0].axis('off')

axes[1, 1].imshow(blur_15)
axes[1, 1].set_title('Blur (15x15)', fontweight='bold')
axes[1, 1].axis('off')

plt.tight_layout()
plt.show()

print("Mayor tamaño del kernel = mayor desenfoque")
````

### 5.2 Detección de Bordes con Canny


````python
# Crear imagen con formas
imagen_formas = np.ones((300, 400, 3), dtype=np.uint8) * 255

# Dibujar formas
cv2.circle(imagen_formas, (100, 100), 50, (0, 0, 255), -1)  # Círculo rojo
cv2.rectangle(imagen_formas, (250, 50), (350, 150), (0, 255, 0), -1)  # Rectángulo verde
cv2.line(imagen_formas, (50, 250), (350, 250), (255, 0, 0), 3)  # Línea azul

# Convertir a escala de grises
imagen_gray = cv2.cvtColor(imagen_formas, cv2.COLOR_BGR2GRAY)

# Aplicar Canny (detección de bordes)
bordes = cv2.Canny(imagen_gray, 50, 150)

# Visualizar
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].imshow(cv2.cvtColor(imagen_formas, cv2.COLOR_BGR2RGB))
axes[0].set_title('Imagen Original con Formas', fontweight='bold')
axes[0].axis('off')

axes[1].imshow(bordes, cmap='gray')
axes[1].set_title('Bordes Detectados (Canny)', fontweight='bold')
axes[1].axis('off')

plt.tight_layout()
plt.show()

print("Canny detecta cambios abruptos de intensidad (bordes)")
````
